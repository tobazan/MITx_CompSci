## SIMULANDO LA EVOLUCION DE UN PACIENTE CON y SIN TRATAMIENTO 

Comparamos primero una simulacion de 300 time steps en la que al paciente NO se le prescribe medicacion y luego una simuacion en la que a partir del time step 150 se le administra la medicacion correcta

En este caso, la probabilidad de que el virus mutee y adquiere resistencia a la medicacion es de 0.005

___

```python
import matplotlib.pyplot as plt 

fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=[10, 6])

ax1.plot(simulationWithoutDrug(100, 1000, 0.1, 0.05, 100))
ax1.set_title("SimpleVirus simulation")
ax1.set_xlabel("Time Steps")
ax1.set_ylabel("Average Virus Population")

resultado = simulationWithDrug(100, 1000, 0.1, 0.05, {"guttagonol": False}, 0.005, 100)
ax2.plot(resultado[0], label='Total Pop')
ax2.plot(resultado[1], label='Resistant Pop')
ax2.set_title("ResistanVirus simulation")
ax2.set_xlabel("Time Steps")
ax2.set_ylabel("Average Virus Population")
ax2.legend(loc = 1)

plt.show()
```

___

Por la bajisima probabilidad de mutacion, la poblacion resistente se mantiene en 0 y a partir de la prescripcion de Guttagonol, el paciente mejora rapidamente.

Ahora, vamos a replicar la simulacion solo que la probabilidad de que el virus mute sera mas elevada de manera que la poblacion resistente varie
___
