## SIMULACION ROBOTS ASPIRADORAS

En el programa **roomba.py** se definieron las clases Posicion, Habitacion y ambos modelos de Robot (Standar y Random Walk)

Ademas, se definio una funcion **runSimulation** que toma como parametros:

- cantidad de robots 
- velocidad de los robots 
- ancho y largo de la habitacion
- % min de cobertura
- nro de simulaciones

y devuelve, el promedio de "pasos" que tomaria aspirar la habitacion hasta ese % de cobertura dado

___

En este notebook vamos a plotear y asi comparar la performance de ambos modelos de robot

```python
from roomba import runSimulation, StandardRobot, RandomWalkRobot
import matplotlib.pyplot as plt
```

```python
def showPlot1():
    num_robot_range = range(1, 11)
    times1 = []
    times2 = []

    for num_robots in num_robot_range:
        times1.append(runSimulation(num_robots, 1.0, 20, 20, 0.8, 20, StandardRobot))
        times2.append(runSimulation(num_robots, 1.0, 20, 20, 0.8, 20, RandomWalkRobot))

    plt.plot(num_robot_range, times1)
    plt.plot(num_robot_range, times2)
    plt.title('PERFORMANCE POR MODELO')
    plt.legend(('STD', 'RAND_WALK'))
    plt.xlabel('NRO DE ROBOTS')
    plt.ylabel('TIME_STEPS')
    plt.show()

showPlot1()
```

___

En el grafico anterior, se compara la performance de ambos modelos en:

- una habitacion de 20 x 20
- cobertura minima de 80% de la superficie
- velocidad de 1.0

A mayor cantidad de robots aspirando en simultaneo la misma habitacion ambos modelos tienen una performance similar. Mientras que el modelo Standar supera al Random Walk cuando a flota de aspiradoras es menor.

___

```python
def showPlot2():
    aspect_ratios = []
    times1 = []
    times2 = []

    for width in [10, 20, 25, 50]:
        height = 300//width
        aspect_ratios.append(float(width) / height)

        times1.append(runSimulation(2, 1.0, width, height, 0.8, 200, StandardRobot))
        times2.append(runSimulation(2, 1.0, width, height, 0.8, 200, RandomWalkRobot))

    plt.plot(aspect_ratios, times1)
    plt.plot(aspect_ratios, times2)
    plt.title('PERFORMANCE POR HABITACION')
    plt.legend(('STD', 'RAND_WALK'))
    plt.xlabel('RATIO ANCHO/LARGO')
    plt.ylabel('TIME_STEPS')
    plt.show()

showPlot2()
```

___

En el grafico anterior, se compara la performance de ambos modelos en:

- flota de 2 aspiradoras
- cobertura minima de 80% de la superficie
- superficie constante 300
- velocidad de 1.0

La forma de la habitacion afecta la performance del modelo Random Walk, mientras que al modelo Standard la forma de la habitacion NO
